```python
import glob
import initials
from langchain_community.document_loaders import TextLoader
from langchain_experimental.text_splitter import SemanticChunker
import os

# Initialize SemanticChunker with embedding and max chunk number
text_splitter = SemanticChunker(initials.embedding)

# Path to the directory containing .txt files
path = "/Users/taha/Desktop/rag/data/Apps & Dienste/*.txt"

# Load all text files except those ending with "_summary.txt"
documents = []
for file_path in glob.glob(path):
    if not file_path.endswith("_summary.txt"):  # Exclude files ending with "_summary.txt"
        loader = TextLoader(file_path)
        documents.extend(loader.load())

# Perform semantic chunking
chunks = []
for doc in documents:
    split = text_splitter.create_documents([doc.page_content])  # Split content into smaller chunks
    # Retain original metadata for each chunk
    for chunk in split:
        chunk.metadata = doc.metadata  # Maintain original metadata
    chunks.extend(split)

# 'chunks' now contains semantically chunked sections of all valid documents
```

```python
import initials

from ragas.testset import TestsetGenerator
#from ragas.testset.evolutions import simple, reasoning, multi_context

generator = TestsetGenerator.from_langchain(llm=initials.model, embedding_model=initials.embedding)

testset = generator.generate_with_langchain_docs(chunks, testset_size=50)
```

```python
testset_df = testset.to_pandas()
```

```python
import pandas as pd

# Save the DataFrame to a CSV file at the specified path
output_file_path = "/Users/taha/Desktop/rag/data/Apps & Dienste/_testset_semantic.csv"
testset_df.to_csv(output_file_path, index=False)  # Dosya yolu güncellendi
```
